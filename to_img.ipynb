```python
import os
import fitz
import shutil
```

```python
def rename_pdf_files(directory):
    # Получаем список всех файлов PDF в директории
    pdf_files = [f for f in os.listdir(directory) if f.endswith('.pdf')]
    
    # Сортируем файлы по имени
    pdf_files.sort()
    
    # Переименовываем файлы
    for i, filename in enumerate(pdf_files):
        # Формируем новое имя файла с ведущими нулями
        new_name = f"{i+1:02d}.pdf"
        old_path = os.path.join(directory, filename)
        new_path = os.path.join(directory, new_name)
        
        # Переименовываем файл
        os.rename(old_path, new_path)
        print(f"Renamed '{filename}' to '{new_name}'")

```

```python
def rename_img(directory):
    # Получаем список всех файлов PDF в директории
    img_files = [f for f in os.listdir(directory) if f.endswith('.jpg')]
    
    # Сортируем файлы по имени
    img_files.sort()
    
    # Переименовываем файлы
    for i, filename in enumerate(img_files):
        # Формируем новое имя файла с ведущими нулями
        new_name = f"img_47_{i+1:02d}.jpg"
        old_path = os.path.join(directory, filename)
        new_path = os.path.join(directory, new_name)
        
        # Переименовываем файл
        os.rename(old_path, new_path)
        print(f"Renamed '{filename}' to '{new_name}'")
```

```python
def file_to_png(file_path, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    # Открываем PDF файл
    try:
        pdf_document = fitz.open(file_path)
        num_pages = len(pdf_document)
        num_digits = len(str(num_pages))  # Количество цифр в номере страницы

        # Проходим по всем страницам PDF и сохраняем их в формате PNG
        for page_number in range(num_pages):
            page = pdf_document.load_page(page_number)
            image = page.get_pixmap()

            # Сохраняем изображение с ведущими нулями в имени файла
            image.save(os.path.join(output_folder, f'img_{os.path.splitext(os.path.basename(file_path))[0]}_{str(page_number+1).zfill(num_digits)}.png'))
            print(f'Сохранено изображение img{str(page_number+1).zfill(num_digits)}.png')
        
        # Закрываем PDF файл
        pdf_document.close()
    except PermissionError:
        print('Доступ к файлу запрещен')
    except Exception as e:
        print(f'Произошла ошибка: {e}')
```

```python
output_folder = 'ank_img'

for file in os.listdir('Anketa'):
    if file.endswith('.pdf'):
        file_path = os.path.join('Anketa', file)
        file_to_png(file_path, output_folder)
```

```python
def change_extension_to_png(folder_path):
    for filename in os.listdir(folder_path):
        # Полный путь к файлу
        file_path = os.path.join(folder_path, filename)

        # Проверяем, является ли файл файлом (а не папкой)
        if os.path.isfile(file_path):
            # Получаем имя файла без расширения и его текущее расширение
            name, ext = os.path.splitext(filename)

            # Игнорируем файлы, которые уже имеют расширение .png
            if ext.lower() != '.png':
                new_file_path = os.path.join(folder_path, name + '.png')

                # Переименовываем файл
                os.rename(file_path, new_file_path)
```
